```python
from PIL import Image
import numpy as np

# Define the mapping
replacement_dict = {
    10: 1,
    11: 2,
    12: 3,
    13: 4,
    16: 5,
    17: 6
}
```

```python
folds = pd.read_csv('folds.csv')
```

```python
os.makedirs('data', exist_ok=True)
for i in range(5):
    print(i)
    os.makedirs('data/'+str(i), exist_ok=True)
    os.makedirs('data/'+str(i)+'/train/rgb/', exist_ok=True)
    os.makedirs('data/'+str(i)+'/train/labels/', exist_ok=True)
    os.makedirs('data/'+str(i)+'/validation/rgb/', exist_ok=True)
    os.makedirs('data/'+str(i)+'/validation/labels/', exist_ok=True)
    
    df = folds[folds['folds'] != i]
    for filename in df['paths']:
        try:
            print(filename)
            
            source = '/home/klimenko/seg_materials/FINAL_DATASETS/COMBINED/RGB/'+filename
            destination = '/home/klimenko/seg_materials/VAL_SEGFORMER/data/'+str(i)+'/train/rgb/'+filename.replace(".jpg", ".png")
            #shutil.copy(source, destination)
            Image.open(source).save(destination.replace(".jpg", ".png"), format='PNG')
            
            
            source = '/home/klimenko/seg_materials/FINAL_DATASETS/COMBINED/MASKS/'+filename
            destination = '/home/klimenko/seg_materials/VAL_SEGFORMER/data/'+str(i)+'/train/labels/'+filename.replace(".jpg", ".png")
            #shutil.copy(source, destination)
            Image.open(source).save(destination, format='PNG')
            
            
            #
            
            image = Image.open(destination)
            image_array = np.array(image)
            
            modified_array = np.full_like(image_array, 0)
            for key, value in replacement_dict.items():
                modified_array[image_array == key] = value
            modified_image = Image.fromarray(modified_array)
            modified_image.save(destination)

            
        except Exception as f:
            print(f, 'ERROR---------------')
            
            
    df = folds[folds['folds'] == i]
    for filename in df['paths']:
        try:
            print(filename)
            
            source = '/home/klimenko/seg_materials/FINAL_DATASETS/COMBINED/RGB/'+filename
            destination = '/home/klimenko/seg_materials/VAL_SEGFORMER/data/'+str(i)+'/validation/rgb/'+filename.replace(".jpg", ".png")
            #shutil.copy(source, destination)
            Image.open(source).save(destination.replace(".jpg", ".png"), format='PNG')
            
            
            source = '/home/klimenko/seg_materials/FINAL_DATASETS/COMBINED/MASKS/'+filename
            destination = '/home/klimenko/seg_materials/VAL_SEGFORMER/data/'+str(i)+'/validation/labels/'+filename.replace(".jpg", ".png")
            #shutil.copy(source, destination)
            Image.open(source).save(destination.replace(".jpg", ".png"), format='PNG')
            
            #
            
            
            image = Image.open(destination)
            image_array = np.array(image)
            
            
            modified_array = np.full_like(image_array, 0)
            for key, value in replacement_dict.items():
                modified_array[image_array == key] = value
            modified_image = Image.fromarray(modified_array)
            modified_image.save(destination)

        
            
        except:
            print('ERROR---------------')
```
